### Create scatter and bubble plots (2 methods)

```python
# 2 methods to create a scatter plot (matlab method)
# 1st: plt.plot()

plt.plot(x_axis,y_axis,'o') #'o' changes plot from a line plot to a scatter plot
```

```python
# 2nd: plt.scatter()
plt.scatter(x_axis,y_axis)
```

```python
# bubble chart:

plt.scatter(x_axis,y_axis,s=y_axis) # s= is the size of each dot
```

```python
# increase the size of the dots, while keeping them proportional:

# one way to do this is by creating a new array of values, and multiplying each value in y_axis by 3 within this array

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
# another way of resizing is to use a "list comprehension" method to replace a for or while loop

# list comprehension syntax: new_list = [*expression* for *item* in *list* if *conditional*]

plt.scatter(x_axis,y_axis,s = [i*3 for i in y_axis])
```

```python
# create scatter and bubble plots - object oriented method

fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis,s = y_axis)
```

### Create pie charts (2 methods)

```python
# create pie charts, matlab method

plt.pie(y_axis,labels = x_axis)
plt.show()
```

```python
# formatting -- add colors, explode July from the pie, and add percentages to each slice

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# create a pie chart - object oriented method

fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show
```

### Chart extras

```python
# add error bar to a line chart
# syntax: plt.errorbar(x,y,yerr = <value>)

# plot line chart of x_axis and y_axis with standard deviation error bars

stdev = statistics.stdev(y_axis)

plt.errorbar(x_axis,y_axis,yerr=stdev)

```

```python
# adding cap to error bars: "capsize = " parameter

plt.errorbar(x_axis,y_axis,yerr=stdev,capsize=3)
```

```python
# plot line chart with error bar, object oriented method

fig, ax = plt.subplots()
ax.errorbar(x_axis,y_axis,yerr=stdev,capsize=3)
plt.show()
```

```python
# add errorbars to a bar chart -- note, you don't need the errorbar function.

# matlab method:

plt.bar(x_axis,y_axis,yerr=stdev,capsize=3)
```

```python
# object oriented method:

fig, ax = plt.subplots()
ax.bar(x_axis,y_axis,yerr=stdev,capsize=3)
```

```python
# change the major ticks - matlab method

import numpy as np #numpy can be used to create arrays or matrices of numbers
plt.barh(x_axis,y_axis) #use horizontal bar plot as an example
plt.xticks(np.arange(0,51,step=5.0)) #np.arange(low range, high range, step size). output of arange is an array between 0 and 50 (0-51 non inclusive) with an increment of 5
plt.gca().invert_yaxis() #use gca() to get current axis and invert it
```

```python
# object oriented method

fig, ax = plt.subplots()
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))
plt.show()
```

```python
# add minor ticks - matlab method:

from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
